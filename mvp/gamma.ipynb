```python

import cv2
import matplotlib.pyplot as plt # plt 用于显示图片
import matplotlib.image as mpimg # mpimg 用于读取图片
import numpy as np
%matplotlib inline

plt.figure(figsize=[60, 60])

img = mpimg.imread('a.png') # 读取和代码处于同一目录下的 lena.png
# 此时 lena 就已经是一个 np.array 了，可以对它进行任意处理
img.shape #(512, 512, 3)
img1 = np.power(img/float(np.max(img)), 1/3)*np.max(img)
img2 = np.power(img/float(np.max(img)), 3)*np.max(img)
plt.subplot(6, 6, 1)
plt.imshow(img) # 原图
plt.subplot(6, 6, 2)
plt.imshow(img1) # 提亮
plt.subplot(6, 6, 3)
plt.imshow(img2) # 变暗

plt.axis('off') # 不显示坐标轴
plt.show()


# img = cv2.imread('a.png',0)



# cv2.imshow('src',img)
# cv2.imshow('gamma=1/1.5',img1)
# cv2.imshow('gamma=1.5',img2)
# cv2.waitKey(0)




# plt.figure(figsize=[6, 6])
# for ii in np.arange(36):
#     plt.subplot(6, 6, ii + 1)
#     plt.imshow(images[ii, ...])
#     plt.axis("off")
# plt.show()

# # 带噪音的图片
# plt.figure(figsize=[6, 6])
# for ii in np.arange(36):
#     plt.subplot(6, 6, ii + 1)
#     plt.imshow(images_noise[ii, ...])
# plt.show()
```
