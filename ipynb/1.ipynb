```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
```

```python
imI = Image.open('./test/I_a.png')
imR = Image.open('./test/R_a.png')
```

```python
imR_array = np.array(imR)/255
imI_array = np.array(imI)/255
# print(imI_array)
gamma = 0.2
im_array  = np.power(imI_array , gamma)

res = np.multiply(imR_array,im_array)
plt.imshow(res)
```
